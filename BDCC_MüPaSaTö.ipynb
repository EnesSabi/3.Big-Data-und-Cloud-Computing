Aufzeigen der Anzahl der fehlenden Werte pro Spalte

```python
missing_values = df.isnull().sum()
print("Anzahl der fehlenden Werte pro Spalte:")
print(missing_values)
```

 Berechnung des Anteils fehlender Daten

```python
missing_percentage = (missing_values / len(df)) * 100
print("\nProzentsatz der fehlenden Daten pro Spalte:")
print(missing_percentage)
```

Visualisierung der Spalten mit mehr als 20% fehlenden Werten

```python
columns_with_high_missing = missing_percentage[missing_percentage > 1].index
print("\nSpalten mit mehr als 1% fehlenden Daten:")
print(columns_with_high_missing)
```

```python
# DepTime zwischen 20:00 und 05:00 Uhr
df['DepTime'] = pd.to_numeric(df['DepTime'], errors='coerce')
df['NightFlight'] = ((df['DepTime'] >= 2000) | (df['DepTime'] <= 500)).astype(int)

# 6 oder 7 fÃ¼r Samstag/Sonntag
df['WeekendFlight'] = df['DayOfWeek'].isin([6, 7]).astype(int)

# DepTime in Stunden extrahieren
df['DepartureHour'] = (df['DepTime'] // 100).astype(float)
```

```python
# 1. Verteilung der NachtflÃ¼ge
# Binary variables
plt.figure(figsize=(10, 6))
sns.countplot(x='NightFlight', data=df)
plt.title('Verteilung von NachtflÃ¼gen (0 = Tagflug, 1 = Nachtflug)')
plt.xlabel('Nachtflug')
plt.ylabel('Anzahl der FlÃ¼ge')
plt.show()
```

```python
# 2. Verteilung der WochenendflÃ¼ge
plt.figure(figsize=(10, 6))
sns.countplot(x='WeekendFlight', data=df)
plt.title('Verteilung von WochenendflÃ¼gen (0 = Wochentag, 1 = Wochenende)')
plt.xlabel('Wochenendflug')
plt.ylabel('Anzahl der FlÃ¼ge')
plt.show()
```

```python
# 3. Verteilung der Abflugstunden
# Continuous variables
plt.figure(figsize=(10, 6))
sns.histplot(df['DepartureHour'].dropna(), bins=24, kde=True)
plt.title('Verteilung der Abflugstunden')
plt.xlabel('Abflugstunde')
plt.ylabel('Anzahl der FlÃ¼ge')
plt.show()
```

```python
# 4. Verteilung der Flugdistanz
plt.figure(figsize=(10, 6))
sns.histplot(df['Distance'].dropna(), bins=30, kde=True)
plt.title('Verteilung der Flugdistanz (Meilen)')
plt.xlabel('Flugdistanz (Meilen)')
plt.ylabel('Anzahl der FlÃ¼ge')
plt.show()
```
